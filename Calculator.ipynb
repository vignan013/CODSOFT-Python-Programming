```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# Initialize history list
history = []

# Function to perform calculation
def calculate(b):
    clear_output(wait=True)
    display(ui)  # Ensure the UI is displayed again after clearing output

    try:
        num1 = float(num1_input.value)
        num2 = float(num2_input.value)
        operation = operation_dropdown.value

        if operation == 'Add':
            result = num1 + num2
        elif operation == 'Subtract':
            result = num1 - num2
        elif operation == 'Multiply':
            result = num1 * num2
        elif operation == 'Divide':
            if num2 != 0:
                result = num1 / num2
            else:
                result = 'Error: Division by zero'
        else:
            result = 'Invalid operation'

        result_label.value = f"Result: {result}"
        history.append(f"{num1} {operation} {num2} = {result}")
        history_label.value = "History:\n" + "\n".join(history)
    except ValueError:
        result_label.value = "Error: Invalid input"

# Creating UI elements
num1_input = widgets.Text(description="Number 1:")
num2_input = widgets.Text(description="Number 2:")
operation_dropdown = widgets.Dropdown(
    options=['Add', 'Subtract', 'Multiply', 'Divide'],
    description='Operation:'
)
calculate_button = widgets.Button(description="Calculate")
result_label = widgets.Label(value="Result:")
history_label = widgets.Label(value="History:")

# Setting up button click event handler
calculate_button.on_click(calculate)

# Displaying the UI
ui = widgets.VBox([num1_input, num2_input, operation_dropdown, calculate_button, result_label, history_label])
display(ui)
```
